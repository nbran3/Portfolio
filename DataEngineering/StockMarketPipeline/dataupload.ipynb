```python
closedf = df['Close']
highdf = df['High']
voldf = df['Volume']
opendf = df['Open']
lowdf = df['Low']
```

```python
closedf
```

```python
table_id = "example.smdata.Close"
```

```python
max_chunks = 10 
counter = 0

while counter < max_chunks:
    t_start = time()
    pandas_gbq.to_gbq(closedf, table_id, project_id=project_id, chunksize=100000, if_exists='append')
    t_end = time()

    print('Inserted another chunk... took %.3f second ' % (t_end - t_start))
    
    counter += 1
    print(f'Processed chunk {counter}/{max_chunks}')

```
