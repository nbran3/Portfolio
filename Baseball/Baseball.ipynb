## One interesting hypothesis I have is that the number of home runs is postively correlated with the number of strike outs.

```python
HR = MLB.groupby(MLB['yearID'])['HR'].sum()
years = list(range(1946, 2016))


plt.xlabel('Year')
plt.ylabel('Total Home Runs')
plt.title('Bar Graph of Total Home Runs per Year')
plt.bar(years, HR, align= 'center')
plt.show()
```

```python
SOA = MLB.groupby(MLB['yearID'])['SOA'].sum()
years = list(range(1946, 2016))

plt.bar(years, SOA, align= 'center', color = 'green')
plt.xlabel('Year')
plt.ylabel('Total Strike Outs')
plt.title('Bar Graph of Total Strike Outs per Year')
plt.show()
```

## The bar graph of strike outs also show that the number of strikeouts has also increased consistently. These bar graphs support my hypothesis, as both variables have increased in values as time went on. This makes sense as batters are trying to swing for the fences more, they more likely they are going to be taking bigger swings, which can often lead to batters missing the ball more due to bad timing of pitches or other factors which could result in more swing and misses.

```python
plt.hist(SOA, 10, edgecolor = 'black',  color = 'green')
plt.xlabel('Total Home Runs')
plt.ylabel('Frequency')
plt.title('Distribution of Strikeouts per Year')
plt.show()
```

```python
corr = HR.corr(SOA)
corr
```

## From the bar graphs and the correlation analysis it appears that the two variables, the number of home runs and the number of strikes in one season probably do share a positive relationship with each other, meaning as one variable goes up, so does another, but correlation does not imply causation and more statistical analyis will need to be done.

```python
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
x = HR.values.reshape(-1,1)
y = SOA.values.reshape(-1,1)

model = LinearRegression()
model.fit(x,y)
x = sm.add_constant(x)
model = sm.OLS(y,x)
results = model.fit()
results.summary()
```

## From the regression analysis we can see that 86% of the variation in the y-variable, the number of Strike Outs per season can be attributed to the number of home runs in a season via using the Least Sqaures method The p-value is 0.000 under the 5% significance level that normally use during hypothesis testing and with a F-stat of 430.5 and its own p-value of something very small (>.05), we can conclude that this model is statistically significant.

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10))


sm.qqplot(HR, line='s', ax=axs[0, 0])
axs[0, 0].set_title('Normality Plot - Home Runs')

sm.qqplot(SOA, line='s', ax=axs[0, 1])
axs[0, 1].set_title('Normality Plot - Strikeouts')

model_hr = sm.OLS(HR, sm.add_constant(SOA)).fit()  
model_soa = sm.OLS(SOA, sm.add_constant(HR)).fit()  
residuals_hr = HR - model_hr.predict(sm.add_constant(SOA))
residuals_soa = SOA - model_soa.predict(sm.add_constant(HR))

sns.residplot(x=HR, y=residuals_hr, ax=axs[1, 0])
axs[1, 0].set_title('Residual Plot - Home Runs')
sns.regplot(x=HR, y=residuals_hr, ax=axs[1, 0], scatter=False)

sns.residplot(x=SOA, y=residuals_soa, ax=axs[1, 1])
axs[1, 1].set_title('Residual Plot - Strikeouts')
sns.regplot(x=SOA, y=residuals_soa, ax=axs[1, 1], scatter=False)

plt.tight_layout()
plt.show()
```

## The normality plots indicate that the data closely aligns with a normal distribution, with only minor deviations. The residual plots show no recognizable pattern, which is a positive indication. It suggests that our model is reliable and fits the data well, as there are no systematic deviations or trends in the residuals.

## In conclusion, the analysis supports the hypothesis that the number of home runs is positively correlated with the number of strikeouts. The data shows a consistent trend of increasing values for both variables, indicating a positive relationship between home runs and strikeouts
