```python
price_data = pd.merge_asof(price_data, fed_data, left_index=True, right_index=True)

price_data
```

```python
price_data.columns = ['price', 'value', 'interest', 'vacancy', 'cpi', 'sp500', 'consumersent', 'supply']

price_data
```

```python
price_data.plot.line(y='price', use_index = True)
```

```python
price_data["adj_price"] = price_data["price"] / price_data["cpi"] * 100
price_data["adj_value"] = price_data["value"] / price_data["cpi"] * 100
```

```python
price_data.plot(y='adj_price', use_index=True)
```

```python
price_data['next_quarter'] = price_data['adj_price'].shift(-13)
```

```python
price_data
```

```python
price_data.dropna(inplace=True)
```

```python
price_data
```

```python
price_data["change"] = (price_data["next_quarter"] > price_data["adj_price"]).astype(int)

price_data
```

```python
price_data["change"].value_counts()
```

```python
predictors = ["price", "sp500", "consumersent","interest", 'supply', "vacancy", "adj_price", "adj_value"]
target = "change"
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import numpy as np

START = 260
STEP = 52

def predict(train, test, predictors, target):
    rf = RandomForestClassifier(min_samples_split=10, random_state=1)
    rf.fit(train[predictors], train[target])
    preds = rf.predict(test[predictors])
    return preds

def backtest(data, predictors, target):
    all_preds = []
    for i in range(START, data.shape[0], STEP):
        train = price_data.iloc[:i]
        test = price_data.iloc[i:(i+STEP)]
        all_preds.append(predict(train, test, predictors, target))
    
    preds = np.concatenate(all_preds)
    return preds, accuracy_score(data.iloc[START:][target], preds)
```

```python
preds, accuracy = backtest(price_data, predictors, target)
```

```python
preds
```

```python
accuracy
```

```python
yearly = price_data.rolling(52, min_periods=1).mean()
yearly
```

```python
yearly_ratios = [p + "_year" for p in predictors]
price_data[yearly_ratios] = price_data[predictors] / yearly[predictors]

price_data
```

```python
preds, accuracy = backtest(price_data, predictors + yearly_ratios, target)
```

```python
accuracy
```

```python
pred_match = (preds == price_data[target].iloc[START:])

pred_match[pred_match == True] = 'green'
pred_match[pred_match == False] = 'red'
```

```python
import matplotlib.pyplot as plt

plot_data = price_data.iloc[START:].copy()

plot_data.reset_index().plot.scatter(x="index", y="adj_price", color=pred_match), price_data.plot(y='adj_price')
```

```python
plot_data_reset = plot_data.reset_index()

fig, ax = plt.subplots()

ax.scatter(plot_data_reset["index"], plot_data_reset["adj_price"], color=pred_match, label='Predicted Price')

ax.plot(price_data.index, price_data['adj_price'], color='black', alpha =.6, label='Actual Price')

ax.set_xlabel('Date')
ax.set_ylabel('Price')
ax.set_title('Predicted vs Actual Housing Price adjusted for Inflation')
ax.legend()

plt.show()
```
