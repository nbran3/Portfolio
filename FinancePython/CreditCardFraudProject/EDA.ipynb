```python
import pandas as pd

df = pd.read_csv('Data Science/creditcardFraud.csv')

df
```

```python
### Checking the structure of the dataset

df.describe
```

```python
### Checking if the dataset has any missing or null values
df.isnull().sum()
```

```python
### Checking the imbalance of the dataset, class is the variable we are trying to predict
df['class'] = df['class'].replace({'no': 0, 'yes': 1})
df['class'].value_counts()
```

```python
import matplotlib.pyplot as plt
num_rows = 6
num_cols = 5
num_bins = 10

fig, axes = plt.subplots(num_rows, num_cols, figsize=(25, 32))

axes = axes.ravel()

for i, col in enumerate(df.columns):
    ax = axes[i]
    ax.hist(df[col], bins=num_bins, alpha=0.7)
    ax.set_title(col)
    ax.grid(True)
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')

for i in range(len(df.columns), num_rows * num_cols):
    axes[i].axis('off')


plt.tight_layout()
plt.show()
```

```python
### Checking to see is there is any correlation between the variables of the dataset
import seaborn as sns
import numpy as np
correlation_matrix = df.corr()
np.fill_diagonal(correlation_matrix.values, -2)

plt.figure(figsize=(20,10))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of all variables')
plt.show()
```
