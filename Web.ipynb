```python
import pandas as pd
import csv
import requests
import time
import os
import matplotlib.pyplot as plt
```

```python

api = 'test'

url = 'https://apps.bea.gov/api/data'

state_codes = ['00000','01000', '02000', '04000', '05000', '06000', '08000', '09000', '10000', '11000', '12000', '13000',
               '15000', '16000', '17000', '18000', '19000', '20000', '21000', '22000', '23000', '24000', '25000',
               '26000', '27000', '28000', '29000', '30000', '31000', '32000', '33000', '34000', '35000', '36000',
               '37000', '38000', '39000', '40000', '41000', '42000', '44000', '45000', '46000', '47000', '48000',
               '49000', '50000', '51000', '53000', '54000', '55000', '56000']

all_data = []

for state_code in state_codes:
    params = {
        'UserID': {api},
        'method': 'GETDATA',
        'datasetname': 'Regional',
        'TableName': 'SAGDP1',
        'Year': '2022',
        'ResultFormat': 'JSON',
        'LineCode': '1',
        'GeoFips': state_code,
    }

    response = requests.get(url, params=params)

    if response.status_code == 200:
        
        data = response.json()

       
        all_data.append(data)

     

    else:
        print('Failed to retrieve data from the API.')
    
    time.sleep(5)




```

```python
df = pd.DataFrame(columns=['Place', '2022Value'])

for data in all_data:
    results = data['BEAAPI']['Results']
    place = results['Data'][0]['GeoName']
    value = results['Data'][0]['DataValue']
    df.loc[len(df)] = [place, value]
    
    
    print(f"State: {place}, Value: {value}")


csv_file = '2022.csv'


df.to_csv(csv_file, index=False)
```

```python

merged_df = pd.merge(df1998,df2002, on='Place')
merged_df = pd.merge(merged_df,df2007, on='Place')
merged_df = pd.merge(merged_df,df2012, on='Place')
merged_df = pd.merge(merged_df,df2017, on='Place')
merged_df = pd.merge(merged_df,df2022, on = 'Place')

merged_df.columns= ['Place','1998 GDP','2002 GDP','2007 GDP','2012 GDP','2017 GDP','2022 GDP']

merged_df.to_csv('AllYearsGDP.csv', index = False)
```
