```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('AllYearsGDP.csv')
```

```python
United_States = ['United States']

intervals = ['1998', '2002', '2007', '2012', '2017','2022','2027']
US = df[df['Place'].isin(United_States)]


interval_columns = ['Place'] + [f'{interval} GDP' for interval in intervals]
US_intervals = US[interval_columns]
USGDP98 = US_intervals['1998 GDP'].sum()
USGDP02 = US_intervals['2002 GDP'].sum()
USGDP07 = US_intervals['2007 GDP'].sum()
USGDP12 = US_intervals['2012 GDP'].sum()
USGDP17 = US_intervals['2017 GDP'].sum()
USGDP22 = US_intervals['2022 GDP'].sum()
USGDP27 = US_intervals['2027 GDP'].sum()
USgdp_values = [USGDP98, USGDP02, USGDP07, USGDP12, USGDP17, USGDP22, USGDP27]

plt.plot(intervals, USgdp_values, marker='o')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.title('GDP of United States over Time')


```
