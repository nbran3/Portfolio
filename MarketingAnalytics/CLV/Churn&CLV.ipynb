Here I am using lambdas to apply scores to each customer depending on wheter or not the meet certain criteria that could make them likely to churn or not. Every customer gets at least a 1, simply for being a customer as there will always be a chance for them to churn, but the following are if the Payment is too high, 109 a month was in the top 25% percentile, so having a payment of 109 a month or more gives a 1. For Months since Last Claim if a customer hasn't made a claim in over the top 25% percentile, which is 23 months they might be at a risk of churning because while they might have not gotten into a car accident, it could suggest that they shopping around or inaction which can lead to non-renweal, which adds one point to the customer churn risk score. For Policy Inception I decided at the top 25% percentile, in this case 71 months, if they have been a customer for almost 6 years, they are not likely to churn and would add one point. Number of Open Complaints is an interesting one, because there is a limited number of data points that actually had open complaints, but since they are so rare, I felt here would be the biggest risk for them to churn as if they felt they recieved bad service or it took to long or whatever it would make them very likley to churn, so if there was even one open complaint would add three points to the churn risk score. The last column is just response to a marketing campaign, albeit not the best indictor, no response could signal inaction which could lead to not renewing.

```python
df['customer_flag'] = df['Customer'].apply(lambda x: 1 if pd.notnull(x) else 0)

df['downgrade_flag'] = df['Monthly Premium Auto'].apply(lambda x: 1 if x >= 109 else 0)
df['inactivity_flag'] = df['Months Since Last Claim'].apply(lambda x: 1 if x >= 23 else 0)
df['age_of_account'] = df['Months Since Policy Inception'].apply(lambda x: 1 if x <= 71 else 0)
df['Open_complaints'] = df['Number of Open Complaints'].apply(lambda x: 3 if x >= 1 else 0)
df['non_response_flag'] = df['Response'].apply(lambda x: 1 if x == 'No' else 0)

df['churn_risk_score'] = (
    df['downgrade_flag'] +
    df['inactivity_flag'] +
    df['Open_complaints'] +
    df['age_of_account'] +
    df['non_response_flag'] +
    df['customer_flag']
)
```

Below I calculate both churn rate and the CLV.

First the Churn Rate goes into a binary 2, 1, or 0 depedning on the score. If a customer scored a 1 through a 3 they were given a not likely to churn status or a 0, if a customer scored a 4 through an 6 they were given a 1 to churn indicating a moderate chance to churn, if a customer got scored a 7 or 8, they were given a two, meaning they were rather likely to churn.

Profit Margin was set at 30%, no any particular reason.

Expected Lifespan is key in calculating CLV, I felt that all customers CLV should be calcualted regardlress of wheter or not if they were very long tenured or brand new customers, which explains the minimum and maxiumum lifespan variables. 

Then calculating CLV is just monthly revenue * expected life_span * profit margin, rather basic but I felt it was good for this dataset because there was only about 9000 customers, below you can see the distributions of both CLV and expected lifespan. More in-depth analysis and visualizations will be in the plotly graphs 

```python
df['churn_rate'] = df['churn_risk_score'].apply(lambda x: 1 if 4 <= x <= 6 else (2 if 7 <= x <= 8 else 0))

df['avg_monthly_revenue'] = df['Monthly Premium Auto']

profit_margin = 0.30  

min_lifespan = 1
max_lifespan = 204 

df['expected_lifespan'] = df['churn_risk_score'].apply(
    lambda x: max_lifespan - (x - 1) * (max_lifespan - min_lifespan) / 6
    if x >= 1 else max_lifespan
)


df['CLV'] = df['avg_monthly_revenue'] * df['expected_lifespan'] * profit_margin

print(df[['Customer', 'avg_monthly_revenue', 'churn_rate', 'expected_lifespan', 'CLV']])
```
